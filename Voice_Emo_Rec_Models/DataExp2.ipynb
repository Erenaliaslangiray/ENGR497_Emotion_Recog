```python
X_db = []
for item in X:
    S = librosa.feature.melspectrogram(y=item, sr=sr)
    k = np.mean(librosa.amplitude_to_db(S, ref=np.max).T,axis=0)
    X_db.append(k)
```

```python
plt.figure(figsize=(14,7))
for i in range (len(X_db)):
    if y[i] == 1:
        plt.plot(X_db[i],"r")
    if y[i] == 2:
        plt.plot(X_db[i],"b")
    if y[i] == 3:
        plt.plot(X_db[i],"g")
    if y[i] == 0:
        plt.plot(X_db[i],"k")
```

```python
import pywt
waveletted = []
for item in X:
    a,b = pywt.dwt(item, 'db1')
    waveletted.append(a)
```

```python
X2_new = []
a_new= []
for i in range(len(waveletted)):
    for item in waveletted[i]:
        if item > 0.001:
            a_new.append(item)
    X2_new.append(a_new)
    a_new = []
```

```python
sample_final = []
sampled_list = []
for i in range(len(X2_new)):
    k = 0
    sums = 0
    for item in X2_new[i]:
        if len(sampled_list) == 256:
            continue
        sums = item + sums
        k = k+1
        if k == int(len(X2_new[i])/256):
            sampled_list.append(sums/k)
            k=0
            sums=0
    sample_final.append(sampled_list)
    sampled_list=[]
```

```python
plt.figure(figsize=(14,7))
for i in range (len(X_db)):
    if y[i] == 1:
        plt.plot(sample_final[i],"r")
    if y[i] == 2:
        plt.plot(sample_final[i],"b")
    if y[i] == 3:
        plt.plot(sample_final[i],"g")
    if y[i] == 0:
        plt.plot(sample_final[i],"k")
```

```python
x_tempo = []
for item in X:
    oenv = librosa.onset.onset_strength(y=item, sr=sr)
    tempogram = librosa.feature.tempogram(onset_envelope=oenv, sr=sr)
    x_tempo.append(np.mean(tempogram.T,axis=1))
```

```python
plt.figure(figsize=(14,7))
for i in range (len(X_db)):
    if y[i] == 1:
        plt.plot(x_tempo[i],"r")
    if y[i] == 2:
        plt.plot(x_tempo[i],"b")
    if y[i] == 3:
        plt.plot(x_tempo[i],"g")
    if y[i] == 0:
        plt.plot(x_tempo[i],"k")
```

```python
specband= []
for item in X:
    k = librosa.feature.spectral_bandwidth(item,sr)
    specband.append(k.T)
```

```python
plt.figure(figsize=(14,7))
for i in range (len(X_db)):
    if y[i] == 1:
        plt.plot(specband[i],"r")
    if y[i] == 2:
        plt.plot(specband[i],"b")
    if y[i] == 3:
        plt.plot(specband[i],"g")
    if y[i] == 0:
        plt.plot(specband[i],"k")
```

```python
acglob = []
for item in X:
    oenv = librosa.onset.onset_strength(y=item, sr=sr)
    tempogram = librosa.feature.tempogram(onset_envelope=oenv, sr=sr)
    ac_global = librosa.autocorrelate(oenv, max_size=tempogram.shape[0])
    ac_global = librosa.util.normalize(ac_global)
    acglob.append(ac_global)
```

```python
plt.figure(figsize=(14,7))
for i in range (len(X_db)):
    if y[i] == 1:
        plt.plot(acglob[i],"r")
    if y[i] == 2:
        plt.plot(acglob[i],"b")
    if y[i] == 3:
        plt.plot(acglob[i],"g")
    if y[i] == 0:
        plt.plot(acglob[i],"k")
```
